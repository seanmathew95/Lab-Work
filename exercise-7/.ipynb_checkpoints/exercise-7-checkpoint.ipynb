```python
#read in the file and make dictionaries of DBH at each census
in_file = open("../../class-assignments/classwork-tues-2-21-17/cocoli.txt")
sizes_94 = {}
sizes_97 = {}
sizes_98 = {}

#get rid of first row
in_file.readline()

#work on all subsequent lines, one at a time
for line in in_file:
    items = line.split('\t')
    tag = items[0]
    s94 = items[4]
    s97 = items[5]
    s98 = items[6]
    
    #populate dictionary
    sizes_94[tag] = float(s94)
    sizes_97[tag] = float(s97)
    sizes_98[tag] = float(s98)

print(sizes_94)
in_file.close()
```

```python
import matplotlib.pyplot as plt

#lets make a histogram of DBHs in 1994
# matplotlib works on lists
sizes_in_1994 = list(sizes_94.values())
plt.hist(sizes_in_1994, log = True)
plt.show()
```

```python
sizes_in_1994 = list(sizes_94.values())
sizes_in_1998 = list(sizes_98.values())
differences_94to98 = []
for tag in sizes_94.keys():
    differences_94to98.append(sizes_98[tag] - sizes_94[tag])


f, (ax1, ax2) = plt.subplots(1, 2, sharex = True, sharey = True)
ax1.hist(sizes_in_1994, log = True)
ax1.set_title("Hist in 94")
ax2.hist(sizes_in_1998, log = True)
ax2.set_title("Hist in 98")
plt.show()
```

```python
plt.hist(differences_94to98, log = True)
plt.show()
```

```python
import matplotlib.path as mpath
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt


fig, ax = plt.subplots()

Path = mpath.Path
path_data = [
    (Path.MOVETO, (1.58, -2.57)),
    (Path.CURVE4, (0.35, -1.1)),
    (Path.CURVE4, (-1.75, 2.0)),
    (Path.CURVE4, (0.375, 2.0)),
    (Path.LINETO, (0.85, 1.15)),
    (Path.CURVE4, (2.2, 3.2)),
    (Path.CURVE4, (3, 0.05)),
    (Path.CURVE4, (2.0, -0.5)),
    (Path.CLOSEPOLY, (1.58, -2.57)),
    ]
codes, verts = zip(*path_data)
path = mpath.Path(verts, codes)
patch = mpatches.PathPatch(path, facecolor='r', alpha=0.5)
ax.add_patch(patch)

# plot control points and connecting lines
x, y = zip(*path.vertices)
line, = ax.plot(x, y, 'go-')

ax.grid()
ax.axis('equal')
plt.show()
```
